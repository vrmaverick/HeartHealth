```python
heart = pd.read_csv('./data/heart.csv')
```

```python
over_50 = heart[heart['age']>50]
```

The following is a plot to see relation of Thal and Heart disease
Thalassemia (thal-uh-SEE-me-uh) is an inherited blood disorder that causes your body to have less hemoglobin than normal. But the representation is not good without the use of Object oriented ploting

```python
fig,ax = plt.subplots()
# ax.scatter()
over_50.plot(kind="scatter",x="thal",y="chol",c="target",ax=ax)
ax.set(xlabel="Thalassemia",ylabel="Cholestrol")
```

This Plot shows to analyze <b>Heart Disease Dataset</b> , I have created a subplot based on the Dataset using object oriented methods of Matplotlib.pyplot which is highly <b>Flexible</b> and <b>Customizable</b> 

<li> The Red line Shows the boundary  after which the data can be classified as outliers +ve and -ve side Calculated using Q3+1.5*IQR and Q!-1.5*IQR where IQR is Inter Quartile Range

<li> The Green line indicate the quartiles Q1 qnd Q3 which helps in visualizing Central tendency


```python
plt.style.use('seaborn-whitegrid')
fig,ax = plt.subplots(nrows=2,
                      ncols=1,figsize = (12,12))
scat = ax[0].scatter(x= over_50["age"],y= over_50["chol"],c=over_50["target"],cmap = 'winter')
ax[0].set(ylabel="Cholesterol",title="Heart Disease and Cholestrol levels")
ax[0].set_ylim([125,375])
ax[0].set_xlim([40,80])
ax[0].set_ylabel("Cholesterol", fontsize=14, color='brown',fontweight = "bold")
# over_50.plot(kind="scatter",x="thal",y="chol",c="target",ax=ax)
# ax.set(xlabel="Thalassemia",ylabel="Colestrol")

#legend method for color combination defination
ax[0].legend(*scat.legend_elements(),title = "Target")

#adding horizontal and vertical line for reference
# ax[0].axhline(over_50["chol"].mean(),linestyle = "--")
# ax[0].axvline(over_50["age"].mean(),linestyle="--")

# IQR
iqr1 = over_50["chol"].quantile(0.75)-over_50["chol"].quantile(0.25)
iqr2 = over_50["age"].quantile(0.75)-over_50["age"].quantile(0.25)

#Quartiles
ax[0].axhline(over_50["chol"].quantile(0.75),color = 'green',linestyle = "--")
ax[0].axvline(over_50["age"].quantile(0.75),color = 'green',linestyle="--")
ax[0].axhline(over_50["chol"].quantile(0.25),color = 'green',linestyle = "--")
ax[0].axvline(over_50["age"].quantile(0.25),color = 'green',linestyle="--")

# Outilers Boundary
ax[0].axhline(over_50["chol"].quantile(0.75)+(1.5*iqr1),color = 'red',linestyle = "--")
ax[0].axvline(over_50["age"].quantile(0.75)+(1.5*iqr2),color = 'red',linestyle="--")
ax[0].axhline(over_50["chol"].quantile(0.25)-(1.5*iqr1),color = 'red',linestyle = "--")
ax[0].axvline(over_50["age"].quantile(0.25)-(1.5*iqr2),color = 'red',linestyle="--");

ax[0].tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)

# Second SubPlot
scat1 = ax[1].scatter(x=over_50["age"],y=over_50["thalach"],c=over_50['target'],cmap="autumn")
ax[1].set_ylim([75,225])
ax[1].set_xlim([40,80])
# ax[1].set_ylabel(style='italic', fontsize=14, color='blue')
ax[1].set(xlabel= "Age",ylabel="Thalach",title = "Heart disease and Thalach levels")
ax[1].legend(*scat1.legend_elements(),title = "Target")
ax[1].set_ylabel("Thalach", fontsize=14, color='brown',fontweight = "bold")
ax[1].set_xlabel("Age", fontsize=14, color='brown',fontweight = "bold")

# IQR

iqr3 = over_50["thalach"].quantile(0.75) - over_50["thalach"].quantile(0.25)

# Quartiles for "thalach"
ax[1].axhline(over_50["thalach"].quantile(0.75), color='green', linestyle="--")
ax[1].axvline(over_50["age"].quantile(0.75), color='green', linestyle="--")
ax[1].axhline(over_50["thalach"].quantile(0.25), color='green', linestyle="--")
ax[1].axvline(over_50["age"].quantile(0.25), color='green', linestyle="--")

# Outlier Boundary for "thalach"
ax[1].axhline(over_50["thalach"].quantile(0.75) + (1.5 * iqr3), color='red', linestyle="--")
ax[1].axvline(over_50["age"].quantile(0.75) + (1.5 * iqr2), color='red', linestyle="--")  # Use the correct variable here
ax[1].axhline(over_50["thalach"].quantile(0.25) - (1.5 * iqr3), color='red', linestyle="--")
ax[1].axvline(over_50["age"].quantile(0.25) - (1.5 * iqr2), color='red', linestyle="--")  # Use the correct variable here

#Figure title
fig.suptitle("Heart Disease Analysis", fontsize = 16,fontweight = "bold");
```

<li> From <b>SubPlot2</b> we can deduce that There is a <b>high Probablity that one will get Heart diseases if the age is below 55 and Thalach or max Heart rate is faster than 160<b>
